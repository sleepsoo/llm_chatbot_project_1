```python
@st.cache_resource
def load_vectorstore():
    embeddings = OpenAIEmbeddings(
        model = "text-embedding-3-small",
        dimensions = 1024
    )
    db = FAISS.load_local("my_faiss_index", embeddings)
    return db

df = load_vectorstore()
```
